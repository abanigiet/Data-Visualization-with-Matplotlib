# Experiment 4: Data Visualization with Matplotlib

**Task 1: Line Plot**

```python
# Import the MATPLOTLIB and NUMPY libraries

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create data
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
# Create a line plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sine Wave')
plt.title('Line Plot of Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```

---

**Task 2: Bar Plot**

```python
# Create data
categories = ['A', 'B', 'C', 'D', 'E']
values = [5, 7, 3, 8, 6]
```

```python
# Create a bar plot
plt.figure(figsize=(10, 6))
plt.bar(categories, values, color='#1EA6B3')
plt.title('Bar Plot of Categories')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

---

**Task 3: Histogram**

```python
# Create data
data = np.random.randn(1000)
```

```python
# Create a histogram
plt.figure(figsize=(10, 6))
plt.hist(data, bins=30, color='purple', alpha=0.7)
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

---

**Task 4: Scatter Plot**

```python
# Create data
x = np.random.rand(100)
y = np.random.rand(100)
```

```python
# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(x, y, color='red', alpha=0.5)
plt.title('Scatter Plot of Random Points')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

---

**Task 5: Subplots**

```python
# Create data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
# Create subplots
plt.figure(figsize=(12, 8))
```

```python
# Subplot 1: Sine Wave
plt.subplot(2, 1, 1)
plt.plot(x, y1, label='Sine Wave', color='blue')
plt.title('Sine Wave')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.legend()
plt.grid(True)

# Subplot 2: Cosine Wave
plt.subplot(2, 1, 2)
plt.plot(x, y2, label='Cosine Wave', color='green')
plt.title('Cosine Wave')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.legend()
plt.grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```
